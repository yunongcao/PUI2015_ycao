## 1. Generate 100 samples of different sizes N (N>10 & N<2000) with the same population mean, including a Normal, a Poisson, a Binomial, a Chi-Squared, a Geometric and a Beta.

```python
from __future__ import print_function
import pylab as pl
import numpy as np
%pylab inline
```

```python
mysizes = map(int, np.linspace(10, 2000, 100))
mymean = 100
```

```python
distb = {'Normal':{}, 'Poisson':{}, 'Binomial':{}, 'Chisq':{}, 'Geometric':{}, 'Beta':{}}  
np.random.seed(seed=3)
for i in mysizes:
    distb['Normal'][i] = np.random.normal(mymean, 10, i)
    distb['Poisson'][i] = np.random.poisson(mymean, i)
    distb['Binomial'][i] = np.random.binomial(mymean, 0.6, i)
    distb['Chisq'][i] = np.random.chisquare(mymean, i)
    distb['Geometric'][i] = np.random.geometric(0.6, i)
    distb['Beta'][i] = np.random.beta(mymean, 6, i)
```

## 2. For each sample plot the sample mean against the sample size N. Describe the behavior you see in the plots - do they look as you expected? why?

```python
means = {}
means['Normal'] = {}
fig_mu_N = pl.figure(figsize=(10,5))
axnormal_mu_n = fig_mu_N.add_subplot(111)
for nn in distb['Normal']:
    means['Normal'][nn] = distb['Normal'][nn].mean()
    axnormal_mu_n.plot(nn, means['Normal'][nn], 'o')
axnormal_mu_n.set_xlabel('Sample Size')
axnormal_mu_n.set_ylabel('Sample Mean')
axnormal_mu_n.plot([min(mysizes), max(mysizes)],[mymean, mymean], 'k')
axnormal_mu_n.grid()
plt.title('Normal: Sample Mean vs Sample Size')
```

### For normal distribution,  sample size generally does not affect the sample mean except for a few discrepancies when sample size is below 100. This is as expected.
***

```python
means['Poisson'] = {}
fig_mu_N = pl.figure(figsize=(10,5))
axpoi_mu_n = fig_mu_N.add_subplot(111)
for nn in distb['Poisson']:
    means['Poisson'][nn] = distb['Poisson'][nn].mean()
    axpoi_mu_n.plot(nn, means['Poisson'][nn], 'o')
axpoi_mu_n.set_xlabel('Sample Size')
axpoi_mu_n.set_ylabel('Sample Mean')
axpoi_mu_n.plot([min(mysizes), max(mysizes)],[mymean, mymean], 'k')
axpoi_mu_n.grid()
plt.title('Poisson: Sample Mean vs Sample Size')
```

### For Poisson distribution,  sample mean tend to converges to the sample mean as the size increases. This is as expected since the distribution expresses the probability of a given number of events occurring in a fixed interval of time and/or space if these events occur with a known average rate and independently of the time since the last event.
* * *

```python
means['Binomial'] = {}
fig_mu_N = pl.figure(figsize=(10,5))
axbi_mu_n = fig_mu_N.add_subplot(111)
for nn in distb['Binomial']:
    means['Binomial'][nn] = distb['Binomial'][nn].mean()    
    axbi_mu_n.plot(nn, means['Binomial'][nn], 'o')
axbi_mu_n.set_xlabel('Sample Size')
axbi_mu_n.set_ylabel('Sample Mean')
axbi_mu_n.plot([min(mysizes), max(mysizes)],[mymean, mymean], 'k')
axbi_mu_n.grid()
plt.title('Binomial: Sample Mean vs Sample Size')
```

### For Binomial distribution,  sample size does not affect the sample mean which is as expected since these samples are drawn from a Binomial distribution with the probability of success where p = 0.6

```python
means['Chisq'] = {}
fig_mu_N = pl.figure(figsize=(10,5))
axchi_mu_n = fig_mu_N.add_subplot(111)
for nn in distb['Chisq']:
    means['Chisq'][nn] = distb['Chisq'][nn].mean()    
    axchi_mu_n.plot(nn, means['Chisq'][nn].mean(), 'o')
axchi_mu_n.set_xlabel('Sample Size')
axchi_mu_n.set_ylabel('Sample Mean')
axchi_mu_n.plot([min(mysizes), max(mysizes)],[mymean, mymean], 'k')
axchi_mu_n.grid()
plt.title('Chisquare: Sample Mean vs Sample Size')
```

### Chisquare distribution generally follows the trend of Poisson distribution, but its discrepancies are smaller because the chi-squared distribution with k degrees of freedom is the distribution of a sum of the squares of k independent standard normal random variables.

```python
means['Geometric'] = {}
fig_mu_N = pl.figure(figsize=(10,5))
axgeo_mu_n = fig_mu_N.add_subplot(111)
for nn in distb['Geometric']:
    means['Geometric'][nn] = distb['Geometric'][nn].mean() 
    axgeo_mu_n.plot(nn, means['Geometric'][nn], 'o')
axgeo_mu_n.set_xlabel('Sample Size')
axgeo_mu_n.set_ylabel('Sample Mean')
axgeo_mu_n.plot([min(mysizes), max(mysizes)],[mymean, mymean], 'k')
axgeo_mu_n.grid()
plt.title('Geometric: Sample Mean vs Sample Size')
```
